## WHAT DO YOU MEAN BY PANDAS.

### ---->Pandas is a Python library used for data manipulation and analysis. It provides powerful tools like DataFrames (tables) and Series (1D arrays) to easily load, clean, and analyze structured data, making it especially useful for tasks involving large datasets.








# Here are some key features of pandas:

## Data Structures:

#### 1.Series: A one-dimensional labeled array, similar to a list or column in a table.
#### 2.DataFrame: A two-dimensional table with labeled rows and columns, perfect for storing tabular data.
Data Handling:

#### 3.Data Cleaning: Easily handle missing values, duplicates, and apply transformations.
#### 4.Filtering and Slicing: Extract subsets of data based on conditions or specific indexes.
Data Analysis:

#### 5.Grouping and Aggregation: Perform group-by operations to summarize data (e.g., mean, sum, counts).
#### 6.Merging and Joining: Combine multiple DataFrames to create unified datasets.
File I/O Operations:

#### 7.Read and Write: Import/export data in various formats, like CSV, Excel, JSON, and SQL databases.
Time Series Support:

Supports handling dates, times, and time-based indexing, which is useful for time series analysis.
#### 8.Integration with Visualization Libraries:

Works seamlessly with Matplotlib and Seaborn to visualize data directly from DataFrames.

```python
df.value_counts()
```

### 7.FIND OUT THE ALL LEAD ACTORS ALPHEBETICALLY IN ASCENDING ORDER.

```python
# Correct usage with a specified column name
df = df.sort_values(by='lead', ascending=True)
df
```

### 8.FIND OUT THE ALL LEAD ACTORS ALPHEBETICALLY IN DESCENDING ORDER.

```python
df=df.sort_values(by="lead",ascending=False)
```

```python
df
```

### 9.FIND OUT THE ALL MOVIES ALPHABETICALLY ASCENDING ORDER

```python
df=df.sort_values(by="movie",ascending=True)
```

### 10.FIND OUT THE ALL MOVIES ALPHABETICALLY IN DESCENDING ORDER.

```python
df=df.sort_values(by="movie",ascending=True)
```

### 11.FIND OUT THE 5 TH LAST MOVIES IN THE DATASETS ALPHABETICALLY ASCENDING ORDER.
